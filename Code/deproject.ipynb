### After checking the `location` column, the **cities** can be obtained from the negative *5th* object in the `location` column

```python
# Load `crime` data
#crime = pd.read_csv('/content/ca_offenses_by_city.csv')

#crime.head()
```

```python
# Split the `address` column to get the `city`.

housing['city'] = housing ['address'].str.split(',').str[-5].str.strip()

housing.head()
```

```python
# Check which `cities` where sampled from the `housing` data.
housing['city'].unique()
```

```python
# Group the data using the `city` column.
# This is important because the `crime` data is based on `cities`.
# `Mean` and `sum` were used for aggregation.
# Some columns such as `longitude` and `latitude` can be taken off.
grouped_data = housing.groupby('city').agg({
    'longitude': 'mean',
    'latitude': 'mean',
    'housing_median_age': 'mean',
    'total_rooms': 'sum',
    'total_bedrooms': 'sum',
    'population': 'sum',
    'households': 'sum',
    'median_income': 'mean',
    'median_house_value': 'mean'
}).reset_index()

grouped_data.head()
```

```python
# Check unique cities from the `crime` data.
crime['City'].unique()
```

```python
# Remove the `3` that come with some of the cities
crime['City'] = crime['City'].str.replace('3', '')

crime['City'].nunique
```

```python
# Check unique values of the `city` column from the `grouped_data`
# to confirm whether there are some of the values from it in the `crime` data
grouped_data['city'].unique()
```

```python
# Rename `City` from the `crime` data to `city`.
# This is needed so that the data can be merged using that column.
crime = crime.rename(columns = {'City': 'city'})

crime.info()
```

```python
# Merge data from `crime` and `grouped_data` on the `city` using `left` to maintain all data from both datasets.
merged_data = pd.merge(crime, grouped_data, on='city', how='left')

merged_data.head()
```

```python
merged_data.tail()
```

```python
# Confirm marging
merged_data.info()
```

```python
# Drop some columns and `NaN`
merged_data = merged_data.drop('Rape (legacy definition)', axis = 1).dropna().reset_index(drop=True)

merged_data.head()
```

```python
merged_data.info()
```

```python
cols_to_rename = {
    'Population': 'population',
    'Violent crime': 'violent_crime',
    'Murder and nonnegligent manslaughter': 'murder',
    'Rape (revised definition)': 'rape_cases',
    'Robbery': 'robbery_crime',
    'Aggravated assault': 'assault_crime',
    'Property crime': 'property_crime',
    'Burglary': 'burglary_crime',
    'Larceny-theft': 'larceny_theft',
    'Motor vehicle theft': 'motor_vehicle_theft',
    'Arson': 'arson',
    'housing_median_age': 'average_housing_age',
    'households': 'total_households',
    'median_income': 'average_income',
    'median_house_value': 'average_house_value',
    'population': 'population_drop'
}
```

```python
merged_data = merged_data.rename(columns = cols_to_rename)

merged_data.info()
```

```python
cols_to_drop = ['population_drop', 'longitude', 'latitude']

merged_data = merged_data.drop(cols_to_drop, axis = 1)

merged_data.info()
```
